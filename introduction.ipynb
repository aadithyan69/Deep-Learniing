```python
import cv2
```

```python
img=cv2.imread('images/images.jpeg')
```

```python
img
#3 dimensional array,here values in the array are pixel values(RGB values)
#use of neural network is to identify these patterns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
#opencv reads image in bdr format by default,so it is having blue tone
#we can convert from bdr to RGB.
```

```python
#to view orginal image in opencv
cv2.imshow('image',img)
cv2.waitKey()#if we do not give value inside bracket then image window will not colse until we close it.(millisecond is given inside waitKey())
cv2.destroyAllWindows()#to close the image window press esc key
```

```python
#convert img to RGB to view orginal image in matplotlib.
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(converted_img)
```

```python
#reading as grayscale image.
grey_img=cv2.imread('images/images.jpeg',cv2.IMREAD_GRAYSCALE)
```

```python
#to view grey_img
cv2.imshow('image',grey_img)
cv2.waitKey()#if we do not give value inside bracket then image window will not colse until we close it.(millisecond is given inside waitKey())
cv2.destroyAllWindows()#to close the image window press esc key
```

```python
#converting already readed image to grayscale image.
grey_img_converted=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
#running the grey_img_converted image.
cv2.imshow('grey_img_converted',grey_img_converted)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#displaying both RGB and grayscale image at the same time.(ie,two image at a time)
cv2.imshow('grey_img_converted',grey_img_converted)
cv2.imshow('images',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img.shape
#(height,width,channel)....channel=3 means RGB,for gray scale channel will not be there.
```

```python
#to save the converted image to local.
cv2.imwrite('images/converted_img.jpg',grey_img_converted)
```

```python
#binary image/black and white image(0-128==black and 129-255==white)
#first we need to convert RGB to grayscale before converting into binary image.

(thresh,img_bw)=cv2.threshold(grey_img_converted,128,255,cv2.THRESH_BINARY)#manul method(it can be done usng an algorithm OTSU)
#we cant predict what the value should be,so we use automatic method.
```

```python
cv2.imshow('binary_image',img_bw)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(thresh)
```

```python
#automatic mode using OTSU
#we are using the automatic mode.
(thresh1,img_bw1)=cv2.threshold(grey_img_converted,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
cv2.imshow('binary_img',img_bw1)
cv2.waitKey()
cv2.destroyAllWindows()
```
