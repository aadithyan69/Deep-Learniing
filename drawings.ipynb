```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('images/omg.jpeg')
img
```

```python
plt.imshow(img)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#RECTANGLE
#pt1--point 1()--coordinates in the graph displaying image.pt2 --point 2

rect_img=cv2.rectangle(img_rgb,pt1=(100,25),pt2=(50,75),color=(255,255,0),thickness=5)#when thickness=-1,rectangle will be a filled rectangle.
plt.imshow(rect_img)
'''
rgb=(0,0,0)==>gives black.

r=0-255--0 means dark red,when it increase color gets lighter
g=0-255--0 means darker and when it inceases it gets lighter
b=0-255--0 means darker and when it inceases it gets lighter

(255,0,0)-red
(255,255,0)-yellow
(255,0,255)-cyan
'''
```

```python
#CIRCLE
cir_img=cv2.rectangle(img_rgb,pt1=(190,40),pt2=(55,115),color=(0,0,0),thickness=-1)#rectangle

cir_img=cv2.circle(img_rgb,center=(100,75),radius=30,color=(0,255,0),thickness=-1)
cir_img=cv2.circle(img_rgb,center=(125,75),radius=30,color=(0,0,255),thickness=-1)
cir_img=cv2.circle(img_rgb,center=(150,75),radius=30,color=(255,255,0),thickness=-1)
plt.imshow(cir_img)
```

```python
# Adding text into image.
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
text_img=cv2.putText(img_rgb,text='luminar',org=(25,150),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=2,color=(255,0,0),thickness=2)
plt.imshow(text_img)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
text_img1=cv2.rectangle(img_rgb,pt1=(250,75),pt2=(50,153),color=(255,255,0),thickness=-1)
text_img1=cv2.circle(img_rgb,center=(210,115),radius=40,color=(255,0,0),thickness=-1)
text_img1=cv2.putText(img_rgb,text='luminar',org=(55,115),fontFace=cv2.FONT_HERSHEY_PLAIN,fontScale=2,color=(0,0,255),thickness=2)
plt.imshow(text_img1)
```
